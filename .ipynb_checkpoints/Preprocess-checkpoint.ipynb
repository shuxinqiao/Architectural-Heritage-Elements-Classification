```python
import os
import random
import shutil

import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader

# set seed
random.seed(10)

train_val_path = 'dataset/train_val/'
train_path = 'dataset/train/'
val_path = 'dataset/val/'

ratio = 0.8
```

```python
# Splits train val sets 
for class_name in os.listdir(train_val_path):
    
    images = os.listdir(train_val_path + class_name)
    train_set = random.sample(images, int(len(images) * ratio))
    val_set = list(set(images) - set(train_set))
    
    print('------ ' + class_name + ' -------')
    print('total:', len(images))
    print('train:', len(train_set))
    print('val:', len(val_set))
    print('ratio:', '%.2f' % (len(train_set)/len(images)))
    
    # Create directories
    try:
        os.mkdir(train_path + class_name)
        os.mkdir(val_path + class_name)
    except FileExistsError:
        pass
    
    # Loop all images in train & val, compare with random set and copy file
    for image in os.listdir(train_val_path + class_name):
        if image in train_set:
            shutil.copy(os.path.join(train_val_path, class_name, image), train_path + class_name)
        elif image in val_set:
            shutil.copy(os.path.join(train_val_path, class_name, image), val_path + class_name)

print('--------------')
print('**** Done ****')
```
